```python
import os
import numpy as np
import scipy.io
```

```python
def location(file_name):
    dic_0 = {'car':0,'david':1,'toy':2}
    dic_1 = {'LFVU':0,'LFVV':1,'LFVY':2}
    dic_2 = {'true':0,'app1':1,'app2':2,'app3':3,'app4':4,'app5':5,'prop':6}
    dic_3 = {'1':0,'3':1,'5':2,'7':3,'10':4,
             '15':5,'25':6,'35':7,'50':8,'80':9,
             '100':10,'120':11,'150':12,'180':13,'200':14}
    
    name_list = file_name.split('_')
    loc_0 = dic_0[name_list[1]]
    loc_1 = dic_1[name_list[0]]
    loc_2 = dic_2[name_list[3]]
    loc_3 = dic_3[name_list[2]]
    
    return loc_0,loc_1,loc_2,loc_3
```

```python
def get_value(file,path):
    file = open(path + '/' + file, 'r') 
    data = file.readlines()
    
    bpp_new = data[5].replace('\n',' ')
    bpp_new2 = bpp_new.split(' ')
    bpp_val = float(bpp_new2[1])
    
    psnr_new = data[7].replace('\n',' ')
    psnr_new2 = psnr_new.split(' ')
    psnr_val = float(psnr_new2[2])
        
    return bpp_val,psnr_val
```

```python
path = "metadata"
filelist = os.listdir(path)
results_bpp = np.zeros((3,3,7,15))
results_psnr = np.zeros((3,3,7,15))

for file in filelist:
#     print (file)
#     print (location(file))
#     print (get_value(file,path))
    
    results_bpp[location(file)[0]][location(file)[1]][location(file)[2]][location(file)[3]] = get_value(file,path)[0]
    results_psnr[location(file)[0]][location(file)[1]][location(file)[2]][location(file)[3]] = get_value(file,path)[1]
    
print (results_bpp)
print (results_psnr)
```
